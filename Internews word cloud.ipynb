```python
stopwords = set(STOPWORDS)
```

```python
internews_wc = WordCloud(
    background_color='lightgrey',
    max_words=2000,
    stopwords=stopwords)

internews_wc.generate(internews)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(internews_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
